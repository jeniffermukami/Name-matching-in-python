**Extracting data from a table using  selenium**

Most times Tables uses tagnames that are tr or td,,represents rows and the data inside the rows.
To locate muktiple elements with the same tag name or class name use **find_elements** instead of **find_element**. When you use find_elements you get a list and in most cases we also have to loop through it to get the data inside the list.

```python
matches = driver.find_elements(By.TAG_NAME,'tr')
```

```python
for match in matches:
    print(match.text)
```

**To store each data inside a variable instead of a single variable**

 Each tag name *tr* has 4 *td*s representing each column.Eg to get only the date, use the xpath:
 
           //tr/td[1]
           since its in the first column
           
    

```python
date = []
home_team = []
score = []
date = []
away_team = []

for match in matches:
    date.append(match.find_element(By.XPATH,'./td[1]').text)
    home_team.append(match.find_element(By.XPATH,'./td[2]').text)
    score.append(match.find_element(By.XPATH,'./td[3]').text)
    away_team.append(match.find_element(By.XPATH,'./td[4]').text) 
    home=match.find_element(By.XPATH,'./td[2]').text



    
    print(home)
```

**Export data to pandas df**

```python
import pandas as pd
```

```python
df =pd.DataFrame({'date':date,'home_team':home_team,'score': score,' away_team': away_team})
```

```python
df.to_csv('matches.csv',index=False)
```

**Selecting elements in Drop downs**

The type of drop down here is a dynamic drop down,Ie.the link doesnt change only the content changes.
We can use weights when the website takes long to load ie.*Use the sleep method*  **time.sleep()**


```python
dropdown = Select(driver.find_element(By.ID,'country'))

#Choosing spain element from the dropdown
"""You can select by index,value or visible text"""
dropdown.select_by_visible_text('Spain')
time.sleep(3)
```
